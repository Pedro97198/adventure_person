IMPORT OF LIBRARIES

```python
import os
import pandas as pd
```

OPEN DE CSV THAT WE HAVE TO USE

```python
person = pd.read_csv("Data/Person.Person.csv")
person_phone = pd.read_csv("Data/Person.PersonPhone.csv")
phone_number_type = pd.read_csv("Data/Person.PhoneNumberType.csv")
email_address = pd.read_csv("Data/Person.EmailAddress.csv")
password = pd.read_csv("Data/Person.Password.csv")
business_entity_contact = pd.read_csv("Data/Person.BusinessEntityContact.csv")
address_type = pd.read_csv("Data/Person.AddressType.csv")
business_entity_adress = pd.read_csv("Data/Person.BusinessEntityAdress.csv")
business_entity = pd.read_csv("Data/Person.BusinessEntity.csv")
business_entity_contact = pd.read_csv("Data/Person.BusinessEntityContact.csv")
contact_type = pd.read_csv("Data/Person.ContactType.csv")
state_province=pd.read_csv("Data/Person.StateProvince.csv")
state_province_country_region=pd.read_csv("Data/Person.vStateProvinceCountryRegion.csv")
country_region= pd.read_csv("Data/Person.CountryRegion.csv")
address = pd.read_csv("Data/Person.Address.csv")
```

CLEAN THE COLUMNS THAT CAN MAKE US ERRORRS

```python
dataframes = [state_province,address,country_region,state_province_country_region,business_entity_adress,address_type,phone_number_type,email_address,password,person_phone,business_entity,contact_type,business_entity_contact]
columnas_a_eliminar = ['ModifiedDate', 'rowguid','Día','Mes','Año','Trimestre']


for df in dataframes:
    df.drop(columns=columnas_a_eliminar, errors='ignore', inplace=True)
```

MAKE THE FIRST DATA FRAME

```python
primerpaso = pd.merge(address_type, business_entity_adress, on='AddressTypeID', how='left')
primerpaso
segundopaso= pd.merge(primerpaso,address, on='AddressID',how='left')
segundopaso
tercerpaso=pd.merge(state_province,segundopaso, on='StateProvinceID',how='left')
tercerpaso
cuartopaso= pd.merge(tercerpaso,state_province_country_region, on='CountryRegionCode', how='left')
cuartopaso
final_address= pd.merge(country_region,cuartopaso,on='CountryRegionCode',how='left')
final_address
```

CHECK IF IS NORMAL THE CUANTITY OF NaN THAT IS IS THE FIRST DATA FRAME

```python
num_nan_segunda_columna = final_address.iloc[:, 2].isnull().sum()
print("El número de valores NaN en la segunda columna de quintopaso es:", num_nan_segunda_columna)
```

SECOND DATA FRAME

```python
primerunion=pd.merge(phone_number_type,person_phone, on='PhoneNumberTypeID', how='left')
primerunion
segundaunion=pd.merge(business_entity,primerunion, on='BusinessEntityID',how='left')
segundaunion
terceraunion=pd.merge(email_address,person,on='BusinessEntityID',how='left')
terceraunion
cuartaunion=pd.merge(terceraunion,password,on='BusinessEntityID',how='left')
cuartaunion
final_person=pd.merge(segundaunion,cuartaunion,on='BusinessEntityID',how='left')
final_person
```

TERCER CONJUNTO DE DATOS

```python
final_contact=pd.merge(contact_type,business_entity_contact,on='ContactTypeID',how='left')
final_contact
```

ANOTHER TRY OF JOIN

```python
prueba=pd.merge(final_person,final_contact,on='BusinessEntityID',how='left')
prueba
```

```python
num_nan_segunda_columna = prueba.iloc[:, 10].isnull().sum()
print("El número de valores NaN en la segunda columna de quintopaso es:", num_nan_segunda_columna)
```

DOWNLOAD DE FINAL DATA 

```python
import os

# Crear la carpeta si no existe
carpeta = "Datos-version1"
if not os.path.exists(carpeta):
    os.makedirs(carpeta)

# Rutas de los archivos CSV
ruta_final_contact = os.path.join(carpeta, 'final_contact.csv')
ruta_final_person = os.path.join(carpeta, 'final_person.csv')
ruta_final_address = os.path.join(carpeta, 'final_address.csv')
ruta_prueba = os.path.join(carpeta, 'prueba.csv')

# Guardar los archivos CSV en la carpeta especificada
final_contact.to_csv(ruta_final_contact, index=False)
final_person.to_csv(ruta_final_person, index=False)
final_address.to_csv(ruta_final_address, index=False)
prueba.to_csv(ruta_prueba, index=False)
```
