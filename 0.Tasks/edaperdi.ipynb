```python
import os
import pandas as pd

carpeta = '/content/'

dataframes = []

for archivo in os.listdir(carpeta):
    if archivo.endswith('.csv'):
        ruta_completa = os.path.join(carpeta, archivo)
        df = pd.read_csv(ruta_completa)
        dataframes.append(df)
```

```python
for i in range(len(dataframes)):
  print(dataframes[i])
```

```python
person = pd.read_csv("Person.Person.csv")
person_phone = pd.read_csv("Person.PersonPhone.csv")
phone_number_type = pd.read_csv("Person.PhoneNumberType.csv")
email_address = pd.read_csv("Person.EmailAddress.csv")
password = pd.read_csv("Person.Password.csv")
business_entity_contact = pd.read_csv("Person.BusinessEntityContact.csv")
```

```python
email_address.columns
```

```python
dataframes = [person, person_phone, phone_number_type, email_address, password, business_entity_contact]
columnas_a_eliminar = ['ModifiedDate', 'rowguid']


for df in dataframes:
    df.drop(columns=columnas_a_eliminar, errors='ignore', inplace=True)
```

```python
print(person)
print(person_phone)
print(business_entity_contact)
```

```python
merged_data = pd.merge(person_phone, phone_number_type, on='PhoneNumberTypeID', how='left')
merged_data = pd.merge(person, merged_data, on='BusinessEntityID', how='left')
merged_data = pd.merge(merged_data, email_address, on='BusinessEntityID', how='left')
```

```python
merged_data = pd.merge(merged_data, password, on='BusinessEntityID', how='left')
final_person = pd.merge(merged_data, business_entity_contact, on='BusinessEntityID', how='left')
```

Empezamos a juntar el segundo dataframe


```python
address_type = pd.read_csv("Person.AddressType.csv")
business_entity_adress = pd.read_csv("Person.BusinessEntityAddress.csv")
business_entity = pd.read_csv("Person.BusinessEntity.csv")

business_entity_contact = pd.read_csv("Person.BusinessEntityContact.csv")
contact_type = pd.read_csv("Person.ContactType.csv")
```

```python
dataframes = [address_type, business_entity_adress,business_entity, business_entity_contact, contact_type]
columnas_a_eliminar = ['ModifiedDate', 'rowguid']

for df in dataframes:
    df.drop(columns=columnas_a_eliminar, errors='ignore', inplace=True)
```

```python
merged_data = pd.merge(business_entity_adress, address_type, on='AddressTypeID', how='left')
merged_data = pd.merge(business_entity, merged_data, on='BusinessEntityID', how='left')
merged_data_1 = pd.merge(business_entity_contact, contact_type, on='ContactTypeID', how='left')
print(merged_data_1)
business_final = pd.merge(merged_data, merged_data_1, on='BusinessEntityID', how='left')
print(business_final)
business_final = pd.merge(business_final, final_person, on='BusinessEntityID', how='left')
```
